# **APIs**  
**API (Application Programming Interface)** is a way for different software applications to communicate with each other. APIs define a set of rules that allow one system to request and receive data from another.  

#### **How APIs Work**  
1. **Client Makes a Request** – A user or another system sends a request to the API (e.g., a mobile app requesting user data).  
2. **API Processes the Request** – The API processes the request, possibly interacting with a database or external service.  
3. **API Sends a Response** – The API returns a response, usually in **JSON** or **XML** format, containing the requested data or an error message.  

For example, when you use a weather app, it sends a request to a weather API, which retrieves the latest weather data and returns it to the app.  

---

### **RESTful API Concepts: HTTP Methods (GET, POST, PUT, DELETE)**  
A **RESTful API** (Representational State Transfer API) follows REST principles, allowing applications to communicate over HTTP using standard methods:  

1. **GET** – Retrieve data  
   - Example: `GET /users` → Returns a list of users  
   - Example: `GET /users/1` → Returns data for user with ID 1  

2. **POST** – Create new data  
   - Example: `POST /users` with `{"name": "Alice", "age": 25}`  
   - Creates a new user and returns a success response  

3. **PUT** – Update existing data  
   - Example: `PUT /users/1` with `{"age": 26}`  
   - Updates the age of the user with ID 1  

4. **DELETE** – Remove data  
   - Example: `DELETE /users/1` → Deletes user with ID 1  

These methods follow **CRUD (Create, Read, Update, Delete)** operations.

---

### Request Parameters
Passing parameters to the RESTful API can be done in different ways.
*   **Query parameters** are the key-value pairs that appear in the URL after a `?`. Multiple query parameters can be included by separating them with an `&`.
```
http://shop.com/products?category=books&promotion=payday&page=2
```
*   **Path parameters** are part of the URL path, They are typically used to represent specific resources.
```
http://shop.com/product/p123
```

## **Flask Application Structure**

Flask is a lightweight Python web framework that is flexible and easy to use for developing web applications, including APIs. The structure of a Flask project can vary depending on the complexity of the application, but there are common components and best practices for organizing a Flask project.

### **Basic Structure of a Flask App**  

A simple Flask application might look like this:

```
/flask_app
    /static
    /templates
    app.py
```

#### **1. app.py**
This is the main file where your Flask app is created and configured. It contains routes and logic for handling HTTP requests.  
A basic example would look like this:

```python
from flask import Flask, jsonify

app = Flask(__name__)

@app.route('/')
def home():
    return jsonify({"message": "Welcome to Flask!"})

if __name__ == '__main__':
    app.run(debug=True)
```

Creating Flask application usually includes the following steps:
1.   Import Flask from the Flask Python package
```python
from flask import Flask
```
2.   Create Flask application with constructor
```python
app = Flask(__name__)
```
Noted that `__name__` is a special variable that holds the name of the current module.
3.   Use decorator `@app.route(PATH)` above function to identify the URL associated with the function. When a request is sent to the URL, the function is called to return the response. The following code depicts that when a request for root path is arrived, return JSON object `{"message": "Welcome to Flask!"}` as the response.

```python
@app.route('/')
def home():
    return jsonify({"message": "Welcome to Flask!"})
```
Noted that we can specify accepted HTTP method(s). For example, the following code shows how to explicitly accept GET method. It receives path parameter `<name>` from request URL.
```python
@app.route('/user/<name>', methods=['GET'])
def home(name):
    return jsonify({"message": f"Welcome {name} to Flask!"})
```

4.   Check if the name of the current module (stored in variable `__name__`) is `__main__`, which means that this script is executed directly not executed as an imported module. If this condition is true, run the Flask application by calling the `app.run()` function. To run in debug mode, call `app.run(debug=True)`

```python
if __name__ == '__main__':
    app.run(debug=True)
```






#### **2. /static**
This folder contains static files like images, JavaScript, CSS, etc. These files are served to the client as they are, and they don't change dynamically.

Example:
```
/static
    /css
        style.css
    /images
        logo.png
    /js
        script.js
```

#### **3. /templates**
Flask uses Jinja2 for templating, which allows you to render HTML templates with dynamic content. This folder typically contains all the HTML files used by the app.

Example:
```
/templates
    index.html
    user_profile.html
```

In a route, you would render an HTML page like this:

```python
from flask import render_template

@app.route('/')
def home():
    return render_template('index.html', name="User")
```

In Flask, **`render_template('html_template', key1=value1, key2=value2)`** command is used to render the specified HTML template and return the output document.
*   `html_template` : HTML file inside the `templates` folder of Flask project.
*   `key1=value1, key2=value2`: Key-value pairs of variables and corresponding values to be injected into the template by the rendering engine.





## **GET Method in APIs**  

### **Characteristics of GET Requests:**  
- **Read-Only**: Does not modify data on the server.  
- **Idempotent**: Multiple identical GET requests will return the same response (if the data hasn't changed).  
- **Cacheable**: Responses can be cached by browsers or intermediary servers to improve performance.  
- **No Request Body**: Data is usually sent in the **URL** (query parameters) rather than in the request body.

### Get Posts
Define `get_posts()` function to be called when receive a request to route `/get_posts`. It performs the following operations.
1.   Retrieve query parameters from the URL of GET request with `request.args`. Then, retrieve the parameter named `'query'`. If not found, return the default value `''` (empty string).
2.   Query MongoDB with `collection.find(filter)`; using the following as  filter for documents as follows:
```python
{"content": {"$regex": search_query, "$options": "i"}}
```
This filter for documents that `content` field matches regular expression (string pattern) in variable `search_query`. The `$options` operator specifies modifiers for the regex to use case-insensitive matching (`i`).
3.   For each matched post, convert its ObjectId in `_id` field to string, and return response in JSON format with HTTP status code 200 (OK).  

```python
search_query = request.args.get('query', '')

# Use a case-insensitive regex search on the 'content' field
matching_posts = list(posts_collection.find({
    "content": {"$regex": search_query, "$options": "i"}
}))

# Convert ObjectId to string for JSON serialization
for post in matching_posts:
    post["_id"] = str(post["_id"])

return jsonify({"matching_posts": matching_posts}), 200
```

````python
from flask import Flask, render_template, jsonify, request
from pyngrok import ngrok
from pymongo import MongoClient
import datetime
from bson import ObjectId
import datetime

app = Flask(__name__)

client = MongoClient("mongodb://localhost:27017/")
db = client["postdb"]
posts_collection = db["posts"]
user = {"id": 1, "name": "Alice"}

@app.route('/')
def home():
    return render_template('template.html', user=user)

@app.route('/add_post', methods=['POST'])
def add_post():
    data = request.get_json()

    user_email = data.get("user_email")
    username = data.get("username")
    content = data.get("content")  # Main content of the post

    if not user_email or not username or not content:
        return jsonify({"error": "Missing required fields"}), 400

    timestamp = datetime.datetime.now().isoformat()

    new_post = {
        "user_email": user_email,
        "username": username,
        "timestamp": timestamp,
        "content": content,
        "comments": []  # Initially, no comments
    }

    result = posts_collection.insert_one(new_post)
    new_post["_id"] = str(result.inserted_id)

    return jsonify({"message": "Post added successfully", "post": new_post}), 201

# GET route: Retrieve posts that contain a specific substring in 'content'
@app.route('/get_posts', methods=['GET'])
def get_posts():
    search_query = request.args.get('query', '')

    # Use a case-insensitive regex search on the 'content' field
    matching_posts = list(posts_collection.find({
        "content": {"$regex": search_query, "$options": "i"}
    }))

    # Convert ObjectId to string for JSON serialization
    for post in matching_posts:
        post["_id"] = str(post["_id"])

    return jsonify({"matching_posts": matching_posts}), 200

@app.route('/add_comment', methods=['POST'])
def add_comment():
    data = request.get_json()

    user_email = data.get("user_email")
    username = data.get("username")
    post_id = data.get("post_id")  # Expecting the _id of the post (as a string)
    comment_text = data.get("comment")

    if not user_email or not username or not post_id or not comment_text:
        return jsonify({"error": "Missing required fields"}), 400

    new_comment = {
        "username": username,
        "user_email": user_email,
        "timestamp": datetime.datetime.now().isoformat(),
        "comment": comment_text
    }

    # Add the new comment to the 'comments' list in the specified post
    result = posts_collection.update_one(
        {"_id": ObjectId(post_id)},
        {"$push": {"comments": new_comment}}
    )

    if result.modified_count == 0:
        return jsonify({"error": "Post not found or comment not added"}), 404

    # Fetch the updated post for return
    updated_post = posts_collection.find_one({"_id": ObjectId(post_id)})
    updated_post["_id"] = str(updated_post["_id"])

    return jsonify({"message": "Comment added successfully", "post": updated_post}), 201

@app.route('/delete_all', methods=['DELETE'])
def delete_all():
    result = posts_collection.delete_many({})
    return jsonify({"message": f"Deleted {result.deleted_count} documents from the posts collection."}), 200

if __name__ == '__main__':
    public_url = ngrok.connect(5000)
    print(f" * ngrok tunnel \"{public_url}\" -> http://127.0.0.1:5000")

    app.run(port=5000)
````
