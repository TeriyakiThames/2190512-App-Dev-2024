## Objectives

1. Be familiar with server installation and configuration
2. Review concept of python and databases

### Server installation and configuration
1. Download Oracle Virtual Box from https://www.virtualbox.org/wiki/Downloads
2. Create a VM using the bridged network

bridgeNetwork.png

3. Install Ubuntu LTS https://ubuntu.com/download/desktop on the new VM
4. Install MongoDB https://www.mongodb.com/docs/manual/tutorial/install-mongodb-on-ubuntu/
*   To run mongodb from command line, forking the process
```
sudo mongod --dbpath ~/data/db --bind_ip 127.0.0.1 --logpath ~/data/db/mongod.log --fork
```
*   Check if MongoDB is Running
```
ps aux | grep mongod
```
*   To stop started instance
```
sudo mongod --dbpath ~/data/db --shutdown
```
5. Install MySQL https://ubuntu.com/server/docs/install-and-configure-a-mysql-server
*   To check MySQL service status
```
service mysql status
```
Use `:q` to exit
*   Change authentication method and root password:
```
sudo mysql -e "ALTER USER 'root'@'localhost' IDENTIFIED WITH mysql_native_password BY '8OQsfBWBP4'; FLUSH PRIVILEGES;"
```
*   Create `test_user`
```
mysql -u root -p8OQsfBWBP4 -e "CREATE USER 'test_user'@'localhost' IDENTIFIED BY 'test_p@ssw0rd';"
mysql -u root -p8OQsfBWBP4 -e "GRANT ALL PRIVILEGES ON *.* TO 'test_user'@'localhost';"
mysql -u root -p8OQsfBWBP4 -e "FLUSH PRIVILEGES;"
```
6. Create virtual environment on your VM
* `sudo apt install python3.12-venv`
* `mkdir virtualenv`
* `python3 -m venv ./virtualenv`  
7. Run in virtual environment. Use `source` command is used to execute commands from `virtualenv/bin/activate` created during the virtual environment creation. Within the python virtual environment `(virtualenv)`, install python packages.
* `source virtualenv/bin/activate`
* `pip install -v pymongo`
* `pip install -v mysql-connector-python`
* `pip install -v flask`
* `pip install -v requests`

(Note) To quit the ongoing python virtual environment, use the following command inside the python virtual environmen.
```
deactivate
```
8. Manage Firewall and all services from flask.
* `sudo ufw allow 5000/tcp`
9. Work with python
*  Within the python virtual environment, to enter python, start the Python interpreter:
```
python
```
* Work with python, just like in colab
* To exit python prompt:
```
exit()
```
